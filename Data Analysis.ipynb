```python
#remove special characters and wiki notes
def remove_special_characters_list(column):
    master_list = []
    try:
        for list_names in column:
            temp_list = []
            for name in list_names:
                new_name = re.sub('\(.*?\)', '', name)
                new_name = re.sub('\<.*?\>', '', new_name)
                new_name = re.sub("[^0-9a-zA-Z\. &+-]+", "", new_name)
                new_name = new_name.strip()
                temp_list.append(new_name)
            master_list.append(temp_list)
    except:
        for list_names in column:
            temp_list = []
            new_name = re.sub('\(.*?\)', '', list_names)
            new_name = re.sub('\<.*?\>', '', new_name)
            new_name = re.sub("[^0-9a-zA-Z\. &+-]+", "", new_name)
            new_name = new_name.strip()
            temp_list.append(new_name)
            master_list.append(temp_list)
    return master_list

col_list = ['Directors_Wiki','Producers_Wiki','Story&ScreenPlayWriters_Wiki','Starring_Wiki',
            'Cinematographers_Wiki','ProductionCo_Wiki','DistributionCo_Wiki','Directors_IMDB', 
            'Writers_IMDB','Producers_IMDB','Cinematographers_IMDB','ProductionCos_IMDB',
            'DistributionCos_IMDB','Starring_IMDB'
           ]
for i in col_list:
    print(i)
    final_df[i] = remove_special_characters_list(final_df[i])
```

```python
def remove_special_money(column):
    master_list = []
    for item in column:
        new_item = re.sub('\(.*?\)', '', str(item))
        new_item = re.sub('\<.*?\>', '', new_item)
        new_item = re.sub("[^₩$€¥©0-9a-zA-Z\. -–]+", "", new_item)
        new_item = re.sub("US", "", new_item)
        new_item = re.sub("&lt;", "<", new_item)
        new_item = re.sub("&gt;", ">", new_item)
        new_item = new_item.replace(u'\xa0', ' ')
        master_list.append(new_item)
    return master_list

col_list = ['Budget_Wiki','BoxOffice_Wiki','Budget_IMDB','FirstWkendGross_IMDB','WorldWideGross_IMDB']
for i in col_list:
    print(i)
    final_df[i] = remove_special_money(final_df[i])
```

```python
#splitting names that were combined into one item of list
def remove_morphed_names(column):
    master_list = []
    for list_names in column:
        temp_list = []
        for name in list_names:
            name = re.sub('  ', ' ', name)
            count_spaces = sum(map(lambda x : 1 if ' ' in x else 0, name))  
            if count_spaces in [0,1,2]:
                temp_list.append(name)
            elif count_spaces == 3:
                temp = [x.start() for x in re.finditer(' ', name)]
                name1 = name[0:temp[1]] 
                name2 = name[temp[1] + 1:]
                temp_list.append(name1)
                temp_list.append(name2)
            elif count_spaces == 5:
                temp = [x.start() for x in re.finditer(' ', name)]
                name1 = name[0:temp[1]] 
                name2 = name[temp[1] + 1:temp[3]]
                name3 = name[temp[3] + 1:]
                temp_list.append(name1)
                temp_list.append(name2)
                temp_list.append(name3)
            elif count_spaces == 7:
                temp = [x.start() for x in re.finditer(' ', name)]
                name1 = name[0:temp[1]] 
                name2 = name[temp[1] + 1:temp[3]]
                name3 = name[temp[3] + 1:temp[5]]
                name4 = name[temp[5] + 1:]
                temp_list.append(name1)
                temp_list.append(name2)
                temp_list.append(name3)
                temp_list.append(name4)
            else:
                temp_list.append(name)
        master_list.append(temp_list)
    return(master_list)

col_list = ['Directors_Wiki','Producers_Wiki','Story&ScreenPlayWriters_Wiki','Starring_Wiki','Cinematographers_Wiki']
for i in col_list:
    print(i)
    final_df[i] = remove_morphed_names(final_df[i])
```

```python
##Imputing Null Values as average combined with other's score
    ### change this to be imputed based on genre average mixed with other person's scores
final_df['Hayley Rating (Imputed)']=final_df['Hayley Rating']
final_df['Zach Rating (Imputed)']=final_df['Zach Rating']

h_avg = final_df.loc[np.invert(final_df['Hayley Rating'].isna())]['Hayley Rating'].mean()
z_avg = final_df.loc[np.invert(final_df['Zach Rating'].isna())]['Zach Rating'].mean()

for i in final_df.loc[final_df['Hayley Rating'].isna()].index:
#     print(new_df.loc[[i]]['Title'].values[0])
    final_df.loc[[i], 'Hayley Rating (Imputed)'] = round((h_avg + 2*final_df.loc[[i]]['Zach Rating'])/3, 0)
    
for i in final_df.loc[final_df['Zach Rating'].isna()].index:
#     print(new_df.loc[[i]]['Title'].values[0])
    final_df.loc[[i], 'Zach Rating (Imputed)'] = round((h_avg + 2*final_df.loc[[i]]['Hayley Rating'])/3, 0)
```
